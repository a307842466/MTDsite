```python
import numpy as np

src = r'/data0/sunzhe/dna-binding/4all/'

idl = [i.replace('\n','') for i in open(r'/data0/sunzhe/dna-binding/4all/test_all_id').readlines()]



for i in idl:
    path = src+'f_all/'+i+'.npy'
    a=np.load(path)
    spd33=[]
    hhm=[]
    pssm=[]
    pssm_hhm=[]
    pssm_spd33=[]
    spd33_hhm=[]
    pssm_hhm_hse=[]
    
    asa=[]
    hse=[]
    che=[]
    angle=[]
    for j in range(0,len(a)):
        temp_s=[]
        temp_h=[]
        temp_p=[]
        t_a=[]
        t_h=[]
        t_c=[]
        t_an=[]

        
        for x in range(0,1):
            t_a.append(float(a[j][x]))
        for x in range(1,3):
            t_h.append(float(a[j][x]))
        for x in range(3,6):
            t_c.append(float(a[j][x]))
        for x in range(6,14):
            t_an.append(float(a[j][x]))
        for x in range(14,34):
            temp_h.append(float(a[j][x]))
        for x in range(34,54):
            temp_p.append(float(a[j][x]))
        spd33.append(temp_s)
        hhm.append(temp_h)
        pssm.append(temp_p)
        asa.append(t_a)
        hse.append(t_h)
        che.append(t_c)
        angle.append(t_an)
        
#         pssm_hhm.append(temp_p+temp_h)
#         pssm_spd33.append(temp_p+temp_s)
#         spd33_hhm.append(temp_h+temp_s)
        
        s=temp_p+temp_h
        s=s+t_a
        s=s+t_h
        s=s+t_c
        #s=s+t_an
        
        pssm_hhm_hse.append(s)
        

    pssm_hhm_asa=np.array(pssm_hhm_hse)
    print(pssm_hhm_asa.shape)

    np.save(src+'f_select_temp2/'+i,pssm_hhm_hse)
```
